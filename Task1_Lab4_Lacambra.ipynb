```
VectorA = [(6,7,8,9)]
VectorB = [(20,22,24,26)]
VectorC = [(15,18,21,24)]
VectorD = [(12,16,20,24)]
VectorE = [(10,15,20,25)]
VectorF = [(10,16,22,28)]

VA = []
VB = []
VC = []
VD = []
VE = []
VF = []

for numbers in VectorA:
    for number in numbers:
        sqrt = number **2
        VA.append(sqrt)

for numbers in VectorB:
    for number in numbers:
        sqrt = number **2
        VB.append(sqrt)

for numbers in VectorC:
    for number in numbers:
        sqrt = number **2
        VC.append(sqrt)     

for numbers in VectorD:
    for number in numbers:
        sqrt = number **2
        VD.append(sqrt) 

for numbers in VectorE:
    for number in numbers:
        sqrt = number **2
        VE.append(sqrt) 

for numbers in VectorF:
    for number in numbers:
        sqrt = number **2
        VF.append(sqrt) 

print('Given:                       ',VectorA)
print('The square of all numbers    ',VA)
print('The sum of all numbers:      ',sum(VA))
print('The square root of the sum:  ',sum(VA)**(1/2))
print('#'*66)
print('')
print('Given:                       ',VectorB)
print('The square of all numbers    ',VB)
print('The sum of all numbers:      ',sum(VB))
print('The square root of the sum:  ',sum(VB)**(1/2))
print('#'*66)
print('')
print('Given:                       ',VectorC)
print('The square of all numbers    ',VC)
print('The sum of all numbers:      ',sum(VC))
print('The square root of the sum:  ',sum(VC)**(1/2))
print('#'*66)
print('')
print('Given:                       ',VectorD)
print('The square of all numbers    ',VD)
print('The sum of all numbers:      ',sum(VD))
print('The square root of the sum:  ',sum(VD)**(1/2))
print('#'*66)
print('')
print('Given:                       ',VectorE)
print('The square of all numbers    ',VE)
print('The sum of all numbers:      ',sum(VE))
print('The square root of the sum:  ',sum(VE)**(1/2))
print('#'*66)
print('')
print('Given:                       ',VectorF)
print('The square of all numbers    ',VF)
print('The sum of all numbers:      ',sum(VF))
print('The square root of the sum:  ',sum(VF)**(1/2))
print('#'*66) 
print("")
```
